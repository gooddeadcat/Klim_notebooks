# Анализ ассортимента детской литературы магазина Читай-Город

Итак, мы отдебажили код и собрали по частям полный (на тот момент) ассортимент детской литературы с сайта магазина Читай-Город.
Теперь объединим все файлы в один датафрейм.

```python
import pandas as pd
```

Так как искомые файлы лежат в другой папке, найдём их с помощью библиотеки os.

```python
import os
```

```python
# Посмотрим, что лежит в папке предыдущего дз.

pp = list(os.walk('C:\\Users\\Klim\\Desktop\\Анализ Данных\\Python\\Klim_notebooks\\Homework\\Parsing_project'))
pp
```

```python
# Создадим список адресов интересующих нас файлов.

kids = []
for three in range(len(pp[0][2])):
    if 'kids' in pp[0][2][three] and not 'final' in pp[0][2][three]:
        kids.append(os.path.join(pp[0][0], pp[0][2][three]))
kids
```

```python
# И создадим датафрейм, объединяющий все эти файлы, используя функцию pd.read_csv на список kids через map.

df = pd.concat(map(pd.read_csv, kids)).reset_index(drop=True)
```

Несмотря на предупреждение, датафрейм создан.

```python
# Сразу удалим задвоения.

df.drop_duplicates(inplace=True)
```

```python
# Переиндексуем.

df = df.reset_index(drop=True)
```

```python
# Выведем основные данные.

df.shape
df.info()
```

```python
# Посмотрим, как это выглядит.

df
```

Итак, у нас на руках датафрейм из 43324 строк и 31 столбца.

```python
# Сохранимся в текущую рабочую директорию.

df.to_csv('parsed_chitay_gorod_kids_final.csv', index=False)
```

Записей с пропущенной ценой всего 0,11%. Но посмотрим на эти записи подробнее.

```python
for j in range(len(list(df.columns))):
    print(list(df.columns)[j], df[df['Цена'].isnull() == True][list(df.columns)[j]].unique())
```

Про эти книги можно сказать следующее:
- они не часть какого-либо цикла;
- их нет (и не предвидится) на сайте и в магазиах;
- среди них нет новинок (вышли не позднее 2020 года);
- об их форме (Формат и Вес) утеряны все сведения.

Выведем среди них отдельно книги с оценкой от 4 и выше.

```python
for j in range(len(list(df.columns))):
    print(list(df.columns)[j], df[(df['Цена'].isnull() == True) &
                                  (df['Рейтинг'] >= 4)][list(df.columns)[j]].unique())
```

Вроде ничего невосполнимого.
Теперь проверим, есть ли в ассортименте другие издания про Простоквашино.

```python
df[df['Название'].str.contains('Простоквашино', case=False)]
```

Изданий предостаточно. С чистой совестью удаляем пропуски по цене.

```python
df.drop(df[df['Цена'].isnull() == True].index, axis=0, inplace=True)
```

Теперь цену и цену со скидкой можно преобразовать в int.

```python
df['Цена'] = df['Цена'].astype('int64')
df['Цена со скидкой'] = df['Цена со скидкой'].astype('int64')
```

Заодно преобразуем ID товара в строку, так как это не число.

```python
df['ID товара'] = df['ID товара'].astype('str')
```

```python
df.info()
```

**Аннотация**

Аннотация не является обязательным признаком для книги, поэтому просто заменим пропуски на '-'.

```python
df['Аннотация'] = df['Аннотация'].fillna('-')
```

**Подкатегории**

Найдём категории пропущенных подкатегорий.

```python
cats = list(df[df['Подкатегории'].isnull() == True]['Категория'].unique())
cats
```

Определим, сколько всего подкатегорий в каждой из выделенных категорий.

```python
df[df['Категория'].isin(cats)].groupby('Категория')['Подкатегории'].nunique()
```

Оказывается, детская религиозная литература вообще не разбита на подкатегории,
а для двух других категорий только некоторые книги не выделены в особую подкатегорию. 

Посчитаем эти книги.

```python
cats2 = dict(df[(df['Категория'].isin(cats)) & 
                (df['Подкатегории'].isnull() == True)].groupby('Категория')['Название'].count())
cats2
```

Определим процент таких книг в каждой выделенной категории.

```python
print('Не выделено в подкатегории, % книг: \n')
for cat in cats2:
    print(cat + ': ' + '{}%'.format(round(cats2[cat] / len(df[df['Категория'] == cat]) * 100, 2)))
```

И, так как ничего страшного тут нет, просто присвоим значение признака 'Категория' признаку 'Подкатегории' каждой такой книги.

Но сначала сохранимся в другой файл.
