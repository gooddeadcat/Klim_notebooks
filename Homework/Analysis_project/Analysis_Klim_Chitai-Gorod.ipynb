**Возрастные ограничения**

Попробуем восстановить недостающие данные о возрастных ограничениях на основе данных о подкатегориях.

Чтобы с подкатегориями было удобнее работать, разделим их и переименуем. Теперь структура будет такая: Категория - Раздел - Подраздел. Признак 'Подкатегории' на всякий случай удалять не будем.

```python
df['Раздел'] = df['Подкатегории'].apply(lambda x: x.split('/')[0])

# Подраздел будет равняться разделу, если после слэша ничего нет.
# Например, для Детской религиозной литературы Категория = Раздел = Подраздел.

df['Подраздел'] = df['Подкатегории'].apply(lambda x: x.split('/')[1] 
                                           if len(x.split('/')) == 2 else x)
```
